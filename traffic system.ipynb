```python
import networkx as nx
import random
```

```python
G = nx.Graph()
```

```python
G.add_edge('A', 'B', weight=5)
G.add_edge('B', 'C', weight=3)
G.add_edge('A', 'C', weight=9)
G.add_edge('C', 'D', weight=1)
```

```python
def update_traffic(G):
    for u, v, data in G.edges(data=True):
        data['weight'] = random.randint(1, 10)
```

```python
update_traffic(G)
```

```python
def find_shortest_path(G, source, target):
    return nx.dijkstra_path(G, source, target, weight='weight')
```

```python
path = find_shortest_path(G, 'A', 'D')
print("Shortest path:", path)
```

```python
import tkinter as tk
from tkinter import messagebox
```

```python
def show_shortest_path():
    source = entry_source.get()
    target = entry_target.get()
    path = find_shortest_path(G, source, target)
    messagebox.showinfo("Shortest Path", f"Shortest path from {source} to {target}: {path}")
```

```python
root = tk.Tk()
root.title("Traffic System")
```

```python
tk.Label(root, text="Source:").grid(row=0, column=0)
entry_source = tk.Entry(root)
entry_source.grid(row=0, column=1)
```

```python
tk.Label(root, text="Target:").grid(row=1, column=0)
entry_target = tk.Entry(root)
entry_target.grid(row=1, column=1)
```

```python
tk.Button(root, text="Find Shortest Path", command=show_shortest_path).grid(row=2, column=0, columnspan=2)
```

```python
root.mainloop()
```

```python
def update_and_refresh():
    update_traffic(G)
    root.after(5000, update_and_refresh)
```

```python
update_and_refresh()
```
